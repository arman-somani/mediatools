# MediaTools Backend Server
Run the cells below in order to start your server with **Unlimited Cloudflare Bandwidth** and **Persistent Google Drive Cookies**!

```python
# 1. Mount Google Drive (So you never lose your cookies!)
from google.colab import drive
drive.mount('/content/drive')
```

```python
%%bash
# 2. Download Code, Install Dependencies, and Start Server
echo "📥 Downloading latest code from GitHub..."
mkdir -p /content/mediatools
cd /content/mediatools
if [ ! -d ".git" ]; then
  git clone https://github.com/arman-somani/mediatools.git .
else
  git pull origin main
fi

echo "🍪 Checking for cookies.txt in Google Drive..."
if [ -f "/content/drive/MyDrive/cookies.txt" ]; then
  cp /content/drive/MyDrive/cookies.txt /content/mediatools/backend/cookies.txt
  echo "✅ Successfully copied cookies.txt from Google Drive!"
else
  echo "⚠️ WARNING: No cookies.txt found in Google Drive!"
  echo "👉 Drop your cookies.txt into the root of your Google Drive to bypass YouTube bot checks permanently!"
fi

echo "⚙️ Installing Node.js and FFmpeg..."
curl -fsSL https://deb.nodesource.com/setup_20.x | sudo -E bash - > /dev/null 2>&1
sudo apt-get install -y nodejs ffmpeg > /dev/null 2>&1

cd backend
echo "📦 Installing Backend Packages..."
npm install > /dev/null 2>&1
npm run build > /dev/null 2>&1

echo "🚀 Starting Express Server..."
nohup npm run start > server.log 2>&1 &
sleep 3
echo "✅ Backend is successfully running in the background!"
```

```python
%%bash
# 3. Start Cloudflare Tunnel (UNLIMITED BANDWIDTH!)
echo "🌐 Starting Cloudflare Tunnel..."
wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64
chmod +x cloudflared-linux-amd64
echo ""
echo "============================================================"
echo "COPY THE LINK BELOW THAT ENDS IN .trycloudflare.com"
echo "AND PASTE IT INTO VERCEL AS YOUR NEXT_PUBLIC_API_URL!"
echo "============================================================"
echo ""
./cloudflared-linux-amd64 tunnel --url http://localhost:5000
```
